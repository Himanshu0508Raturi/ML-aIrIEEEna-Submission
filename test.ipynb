### Importing Libraries

```python
import joblib
import numpy as np
import pandas as pd
```

### Importing model

```python
rf_model = joblib.load('rf_model.pkl')
```

### Importing dataset

```python
test_dataset = pd.read_csv("TEST.csv")
```

### Separating ID and feature from the dataset

```python
test_id = test_dataset["ID"]
x_test = test_dataset.drop("ID", axis=1)
```

```python
x_test.head()
```

### Model classification on TEST.CSV file dataset

```python
test_predictions = rf_model.predict(x_test)
```

### Saved the final output on FINAL.csv file

```python
submission = pd.DataFrame({
    "ID": test_id,
    "CLASS": test_predictions
})
submission.to_csv("FINAL.csv", index=False)
```
